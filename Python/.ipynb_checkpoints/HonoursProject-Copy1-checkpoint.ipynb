```python
# Imports
import math as m
import numpy as np
import matplotlib.pyplot as plt
import holoviews as hv
import pylab as pl
from pylab import exp,cos,sin,pi,tan, pi
%matplotlib inline
hv.notebook_extension()
```

```python
def mises(a,k,ref,x):
    ''' basic von mises function
    Inputs
    -----------------------
    - a: determines magnitude
    - k: determines width (low k is wider width)
    - ref: reference angle
    - x: input angle
    '''
    return a*exp(k*(cos(2*(ref-x))))
```

```python
# Tests
k = 0.25;
A = 3;
print (mises(k,A,0.5,0.7))


x = np.arange(-2*pi, 2*pi, 0.1);
VMs = mises(k,A,x,0.3)
plt.plot(x,VMs)
plt.show()

points = [i for i in np.arange(0, pi, 0.1)]
points = [(points[i], mises(k,A,points[i],pi)) for i in range(len(points))]
pair = hv.Curve(points)
pair
```

```python
def mises_curve(a,k,angle_neuron):
    points = [i for i in np.arange(0, pi, pi/angle_neuron[1])]
    points = [(points[i], mises(k,A,points[i],angle_neuron[0])) for i in range(len(points))]
    curve = hv.Curve(points)
    return curve
```

```python
angle_neurons = [(angle,neurons) for angle in np.arange(0, pi, 0.1) for neurons in np.arange(1,20,1)]
data = {angle_neuron:(mises_curve(A,k,angle_neuron))
        for angle_neuron in angle_neurons}
```

```python
levels = hv.HoloMap(data, kdims=['Angle', 'Neurons']).collate()
levels
```
